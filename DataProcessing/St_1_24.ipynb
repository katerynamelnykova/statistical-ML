# Лабораторна робота №1

## Крок 1: 
```
Об’єднати навчальний набір даних (train/X_train.txt – 7352 рядки, 561 стовпець) і тестовий набір даних (test/X_test.txt – 2947 рядків, 561 стовпець), щоб створити один набір даних (10299 рядків, 561 стовпець).
```

````python
import pandas as pd

def load_data(file_path):
    return pd.read_csv(file_path, delim_whitespace=True, header=None)


X_train = load_data("Dataset/train/X_train.txt")
X_test = load_data("Dataset/test/X_test.txt")
X_data = pd.concat([X_train, X_test], axis=0).reset_index(drop=True)
````

````python
X_data.head()
````

````python
X_data.shape
````

## Крок 2:
```Належним чином позначає набір даних, створений на кроці 1, описовими назвами змінних (стовпців) (ви можете призначити назвами стовпців із файлу features.txt).```

````python
features = load_data("Dataset/features.txt")[1].tolist()
X_data.columns = features
````

````python
X_data.head()
````

## Крок 3:

```
Отримати лише вимірювання (стовпці) середнього значення (mean) та стандартного відхилення (std) для кожного вимірювання з набору даних, створеного на кроці 1 (перегляньте файли features_info.txt і features.txt, щоб визначити відповідні стовпці), щоб створити новий набір даних (10299 рядків, 66 стовпців).
```

````python
selected_columns = [col for col in features if "mean()" in col or "std()" in col]
X_selected = X_data[selected_columns]


X_selected.head()
````

````python
X_selected.shape
````

## Крок 4:

```Об’єднує навчальний набір даних міток (train/y_train.txt – 7352 рядки, 1 стовпець) і тестовий набір даних міток (test/y_test.txt – 2947 рядків, 1 стовпець), щоб створити один набір даних (10299 рядків, 1 стовпець)```

````python
y_train = load_data("Dataset/train/y_train.txt")
y_test = load_data("Dataset/test/y_test.txt")
y_data = pd.concat([y_train, y_test], axis=0).reset_index(drop=True)
y_data.columns = ["Activity"]

y_data.shape
````

## Крок 5:

```Використати описові назви дій для іменування дій у наборі даних, створеному на кроці 4 (замість числового слід використовувати рядковий тип даних із описовим значенням).```

````python
activity_labels = load_data("Dataset/activity_labels.txt").set_index(0)[1].to_dict()
y_data["Activity"] = y_data["Activity"].map(activity_labels)

y_data
````

## Крок 6:
```Об’єднати набір даних, створений на кроці 3 (10299 рядків, 66 стовпців), і набір даних, створений на кроці 4 (10299 рядків, 1 стовпець), щоб створити один набір даних (10299 рядків, 67 стовпців).```

````python
final_data = pd.concat([X_selected, y_data], axis=1)

print(final_data.shape)
final_data.head()
````

## Додатковий крок:
```Конвертація отриманих результатів у .csv```

````python
final_data.to_csv("output.csv", index=False)
````
